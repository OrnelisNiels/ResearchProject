```python
import matplotlib.pyplot as plt

def plot_map_comparison(images, map_scores):
    plt.figure(figsize=(12, 6))
    plt.plot(images, map_scores, marker='o', label='mAP scores')

    # Add values to the line with more spacing
    for i, (x, y) in enumerate(zip(images, map_scores)):
        # Skip the first value
        if i == 0:
            continue
        plt.text(x, y + 0.5, f'{y:.2f}%', ha='right', va='bottom', fontsize=12, color='black')

    # Threshold line
    plt.axhline(y=46.6, color='r', linestyle='--', label="mAP real data with 1500 images")
    plt.xticks(images)
    plt.xlabel('Number of Images')
    plt.ylabel('mAP')
    plt.title('mAP vs Number of Images')
    plt.legend()
    plt.show()

# Example data
images = [50, 100, 200, 300, 500, 750, 1000, 1250, 1500]
map_scores = [17.5, 17.5, 24.5, 30.8 ,40.2, 44.8, 43.8, 47.3, 47.9]

# Call the function with your data
plot_map_comparison(images, map_scores)
```
